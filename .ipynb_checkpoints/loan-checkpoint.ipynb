```python
def adjust_year(date_str):
    if isinstance(date_str, str):
        month, year = date_str.split('-')
        year = int(year)
    
        if year > 23:
            year += 1900
        else:
            year += 2000
    
        return f'{year}-{month}'
    else:
        return None

# df_final_copy['issue_d'] = df_final_copy['issue_d'].apply(adjust_year)
# df_final_copy['issue_d'] = pd.to_datetime(df_final_copy['issue_d'], format='%Y-%b')
```

```python
def per_to_int(s):
    if isinstance(s, str) and '%' in s:
        return float(s.replace('%', ''))
    elif isinstance(s, (int, float)):
        return s
    return float('nan')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_final_copy['loan_amnt'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Loan Amounts')
plt.xlabel('Loan Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_final_copy['int_rate_in_percent'], bins=30, color='orange', edgecolor='black')
plt.title('Distribution of Interest Rates')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df_final_copy['grade'].value_counts().sort_index().plot(kind='bar', color='green', edgecolor='black')
plt.xticks(rotation=0)
plt.title('Loan Grades Distribution')
plt.xlabel('Grade')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df_final_copy['home_ownership'], color='red', edgecolor='black')
plt.title('Home Ownership Distribution')
plt.xlabel('Home Ownership')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_loan_interest_by_grade = df_final_copy.groupby('grade', observed=False).agg(
    avg_loan_amt=('loan_amnt', 'mean'),
    avg_interest_rate=('int_rate_in_percent', 'mean')
)

fig, ax1 = plt.subplots(figsize=(10, 9))
ax2 = ax1.twinx()

grades = avg_loan_interest_by_grade.index 
avg_loan_amt = avg_loan_interest_by_grade['avg_loan_amt']
avg_interest_rate = avg_loan_interest_by_grade['avg_interest_rate']

ax1.plot(grades, avg_loan_amt, color='blue', marker='o', label='Avg Loan Amount')
ax1.set_xlabel('Grade')
ax1.set_ylabel('Avg Loan Amount ($)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

ax2.plot(grades, avg_interest_rate, color='orange', marker='o', label='Avg Interest Rate')
ax2.set_ylabel('Avg Interest Rate (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

fig.suptitle('Average Loan Amount and Interest Rate by Grade')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```
